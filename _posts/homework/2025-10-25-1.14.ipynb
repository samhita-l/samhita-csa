Popcorn Hack #1     

```java
// Define the Dog class with a breed property and a method that prints "My dog <name> is a <breed>!"
public class Dog {
    private String name;
    private String breed;

    public Dog(String name, String breed) {
        this.name = name;
        this.breed = breed;
    }

    public void printBreed() {
        System.out.println("My dog " + name + " is a " + breed + "!");
    }
}

// Create and run a Dog example
public class RunDog {
    public static void main(String[] args) {
        Dog myDog = new Dog("Buddy", "Labrador");
        myDog.printBreed();
    }
}

RunDog.main(null);
```

Popcorn Hack #2

```java
public class Counter {
    private int count;

    public Counter() {
        this.count = 0;
    }

    // addition method
    public void add(int n) {
        count += n;
    }

    // subtraction method
    public void subtract(int n) {
        count -= n;
    }

    // multiplication method
    public void multiply(int n) {
        count *= n;
    }

    // division method (guards against divide by zero)
    public void divide(int n) {
        if (n == 0) {
            System.out.println("Cannot divide by zero. Operation skipped.");
            return;
        }
        count /= n;
    }

    // helper to print current count
    public void printCount() {
        System.out.println("Current count: " + count);
    }
}

// Use it:
public class RunCounter {
    public static void main(String[] args) {
        Counter c = new Counter();

        c.add(10);        // addition
        c.printCount();

        c.subtract(3);    // subtraction
        c.printCount();

        c.multiply(4);    // multiplication
        c.printCount();

        c.divide(5);      // division
        c.printCount();

        // demonstrate divide by zero handling
        c.divide(0);
    }
}

RunCounter.main(null);
```

Bonus Hack

```java
// Define the Joke class
class Joke {
    private String setup;
    private String punchline;

    public Joke(String setup, String punchline) {
        this.setup = setup;
        this.punchline = punchline;
    }

    public void tellJoke() {
        System.out.println(setup);
        System.out.println(punchline);
    }
}

// Create and run a joke
public class RunJoke {
    public static void main(String[] args) {
        Joke myJoke = new Joke("Why don’t programmers like nature?",
                               "Because it has too many bugs!");
        myJoke.tellJoke();
    }
}

RunJoke.main(null);
```
